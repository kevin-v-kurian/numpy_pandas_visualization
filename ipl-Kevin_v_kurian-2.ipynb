# 9.	Has Toss-winning helped in winning matches?

```python
matches.groupby(matches.toss_winner==matches.winner).count()
```

```python
tosswin =( matches.toss_winner == matches.winner )

plt.figure(figsize=(15,5))

sns.countplot(tosswin)

plt.show()
```

we can see there is a co-relation between winning the toss and winning the match favouring the toss winners
